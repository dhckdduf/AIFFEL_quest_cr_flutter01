```python
# 파일 읽기모드로 열기
f = open("/content/drive/MyDrive/Avengers.txt", 'r')
data = f.read()
# print(data)

#1. 소문자 변환
lower_data=data.lower()

#2. 기호 제거
import re
eng_str=re.sub(r"[^a-zA-Z\s]","", lower_data)
# print(eng_str)

#3. 단어 단위로 분리
words=re.split(r'\s+',eng_str)
# print(words)

#4. 빈도 수 계산하여 내림차순으로 딕셔너리 저장
import collections
word_count=collections.Counter(words)
print(word_count)


f.close()
```
