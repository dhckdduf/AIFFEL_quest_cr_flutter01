```python
#프로젝트2 load_wine

import sklearn

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

wine=load_wine()
wine_data=wine.data
wine_data.shape           #178개의 데이터가 13개의 정보를 가지고 있음
wine_feature=wine.feature_names
wine_label=wine.target
wine.target_names          #target_name이 class_0, class_1, class_2임

import pandas as pd
wine_df=pd.DataFrame(wine_data, columns=wine_feature)
wine_df["label"]=wine_label             #label 칼럼을 추가하여 데이터프레임에 저장
wine_df
```

```python
# train, test 데이터 분리

X_train, X_test, y_train, y_test = train_test_split(wine_data, wine_label, test_size=0.2, random_state=7)
```

```python
#Decision Tree

from sklearn.tree import DecisionTreeClassifier
decision_tree=DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train,y_train)
y_pred=decision_tree.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
#Random Forest로 학습
from sklearn.ensemble import RandomForestClassifier
random_forest=RandomForestClassifier(random_state=32)
random_forest.fit(X_train,y_train)
y_pred=random_forest.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
#SVM으로 학습
from sklearn import svm
svm_model=svm.SVC()

svm_model.fit(X_train,y_train)
y_pred=svm_model.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
#SGD으로 학습
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()

sgd_model.fit(X_train,y_train)
y_pred=sgd_model.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
#Logistic Regression으로 학습
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()

logistic_model.fit(X_train,y_train)
y_pred=logistic_model.predict(X_test)
print(classification_report(y_test,y_pred))
```
